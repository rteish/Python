```python
#simulating coin flips 
import numpy as np

flips = np.random.choice(['heads', 'tails'], size=10000)
p_heads = np.sum(flips == 'heads') / 10000
print(f"emperical probability of heads: {p_heads}")
```

```python
#expected value of dice roll
import numpy as np
values = np.arange(1, 6)
probs = np.ones(5) / 6

np.sum(values * probs)
```

```python
values = [1, 2, 3, 4, 5, 6]
probs = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

expected_value = sum(x * p for x, p in zip(values, probs))
print("Expected value:", expected_value)
```

```python
#probability density function

import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

SQRT_2PI = math.sqrt(2 * math.pi)

def normal_pdf(x: float, mu: float = 0, sigma: float = 1) -> float:
    return (np.exp(- (x - mu) ** 2 / (2 * sigma ** 2))) / (SQRT_2PI * sigma)
```

```python
#calculate the probability of a random variable, lies between -1 and 1 in a 
#standard normal distribution
from scipy.stats import norm

probability = norm.cdf(1) - norm.cdf(-1)
print(f"Probability: {probability}")
```
