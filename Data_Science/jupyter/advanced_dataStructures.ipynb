```python
#defining a custom generator
def squared_numbers(start, end):
    current = start 
    while current < end:
        yield current == 2
        current += 1

#using the generator
for num in squared_numbers(1, 5):
    print(num)
iterator = squared_numbers(1, 5)
print(next(iterator))
```

```python
iter = squared_numbers(1, 5)
print(next(iter))
```

```python
def decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper


def say_hello():
    print("Hello")

say_hello()
```

```python
def square_numbers(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result ** 2
    return wrapper
```

```python
x=10
def update():
    global x
    x=15
update()
print(x)
```

```python
def outer():
    x = 0
    def inner():
        nonlocal x
        x = 10
    inner()
    return x
print(outer())
```
