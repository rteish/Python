```python
import pandas as pd
data = pd.DataFrame({
    "Student": ["A", "A", "B", "B"],
    "Subject": ["Math", "Science", "Math", "Science"],
    "Marks": [80, 85, 75, 90]
})
print(data)
pivot_table = data.pivot(index="Student", columns="Subject", values="Marks")
print(pivot_table)
```

```python
melted = data.melt(
    id_vars=["Student"],
    value_vars=["Subject", "Marks"],
    var_name="Type",
    value_name="Value"
)
print(melted)
```
