# MongoDB Database Operations
This notebook demonstrates how to connect to a MongoDB database (or use a mock if not available), insert data, and query it.

```python
from pymongo import MongoClient
from pymongo.errors import ServerSelectionTimeoutError, ConnectionFailure
import mongomock

# Try connecting to local MongoDB, fallback to Mock if it fails
try:
    print("Attempting to connect to local MongoDB server...")
    client = MongoClient('mongodb://localhost:27017/', serverSelectionTimeoutMS=2000)
    client.server_info() # Trigger connection check
    print("SUCCESS: Connected to local MongoDB server.")
except (ServerSelectionTimeoutError, ConnectionFailure):
    print("WARNING: Could not connect to local MongoDB server.")
    print("Falling back to Mongomock (in-memory database) for demonstration.")
    client = mongomock.MongoClient()
    
db = client['test_database']
collection = db['users']
```

```python
# Insert one document
print("Inserting one document...")
result = collection.insert_one({
    'name': 'John Doe',
    'age': 30,
    'city': 'New York'
})
print(f"Inserted document with ID: {result.inserted_id}")
```

```python
# Insert multiple documents
print("Inserting multiple documents...")
docs = [
    {'name': 'Jane Smith', 'age': 25, 'city': 'Los Angeles'},
    {'name': 'Mike Johnson', 'age': 35, 'city': 'Chicago'},
    {'name': 'Sarah Williams', 'age': 28, 'city': 'San Francisco'}
]
result_many = collection.insert_many(docs)
print(f"Inserted {len(result_many.inserted_ids)} documents")
```

```python
# Query and display all documents
print("Current documents in collection:")
for doc in collection.find():
    print(doc)
```
